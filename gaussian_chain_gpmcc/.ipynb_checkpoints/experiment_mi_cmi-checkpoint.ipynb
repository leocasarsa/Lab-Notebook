```python
# Plot mi x num_samples for each state color coded
# Plot time x num_samples

def plot_mi_samples_time(all_mi, all_time_mi, sample_list, mi_true):
    mi_arr = np.array(all_mi)
    n_states = mi_arr.shape[1]
    x = sample_list 
    # Figure initialization
    fig, ax = plt.subplots (2, 1, sharex= True)
    fig.set_size_inches(10,7)
    jet = plt.get_cmap('jet') 
    cNorm  = colors.Normalize(vmin=0, vmax=n_states)
    scalarMap = cm.ScalarMappable(norm=cNorm, cmap=jet)
    # Plotting mi for each state per num_samples
    for i_state in range(n_states):
        ax[0].plot(x, mi_arr[:,i_state], color=scalarMap.to_rgba(i_state), lw=3)
    ax[0].axhline(mi_true, c='k', ls='--', label = 'Analytical')
    ax[0].set_ylabel('$MI(X;Z)$', fontsize = 18)
    ax[0].legend()
    # Plotting time per num_samples
    ax[1].plot(x, all_time_mi, 'ko')
    ax[1].set_ylabel('Running time', fontsize = 18)
    ax[1].set_xlabel('Number of Samples')
```
